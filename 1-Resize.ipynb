```python
# -*- coding: utf-8 -*-
import numpy as np
import cv2
import os
import time
```

## Resize images in all dirctories

```python
def resize_image(file, org_dir, target_dir, overwrite = False):
    if not overwrite and os.path.exists(os.path.join(target_dir,file)):
        return
    img = cv2.imdecode(np.fromfile(os.path.join(org_dir,file),dtype=np.uint8),-1)
    #print(img.shape)
    if img.shape[0] > 4000:
        img = cv2.resize(img, (int(img.shape[1] / 5), int(img.shape[0] / 5)))
    elif  img.shape[0] > 3000:
        img = cv2.resize(img, (int(img.shape[1] / 4), int(img.shape[0] / 4)))
    else:
        img = cv2.resize(img, (int(img.shape[1] / 3), int(img.shape[0] / 3)))
        
    cv2.imencode('.jpg',img)[1].tofile(os.path.join(target_dir,file))
def recu_dir(org_dir, cur_dir):
    if not os.path.exists(cur_dir):
        os.mkdir(cur_dir)
    for file in os.listdir(org_dir):
        if os.path.isdir(os.path.join(org_dir,file)):
            abspath = os.path.join(org_dir,file)
            if not os.path.exists(os.path.join(cur_dir, file)):
                os.mkdir(os.path.join(cur_dir, file))
            print('Go into ', os.path.join(cur_dir, file))
            recu_dir(abspath, os.path.join(cur_dir, file))
        else:
            if file[-4:] == '.JPG':
                resize_image(file, org_dir, cur_dir)
            #if file[-4:] == '.MP4':
            #    get_image_from_video(file, org_dir, cur_dir)
```
